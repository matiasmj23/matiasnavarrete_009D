```python
class Vehiculo:


 def __init__(self, patente, dueno, multas, certificado):
    self.patente = patente
    self.dueno = dueno
    self.multas = multas
    self.certificado = certificado

class RegistroVehiculos:

 def __init__(self):
    self.vehiculos = []

def agregar_vehiculo(self, vehiculo):
    self.vehiculos.append(vehiculo)

def buscar_vehiculo(self, patente):
    for vehiculo in self.vehiculos:
      if vehiculo.patente == patente:
        return vehiculo
    return None

def buscar_multas(self, patente):
    vehiculo = self.buscar_vehiculo(patente)
    if vehiculo is not None:
      return vehiculo.multas
    return None


def buscar_dueno(self, patente):
    vehiculo = self.buscar_vehiculo(patente)
    if vehiculo is not None:
      return vehiculo.dueno
    return None

def imprimir_certificado(self, patente):
    vehiculo = self.buscar_vehiculo(patente)
    if vehiculo is not None:
      print(vehiculo.certificado)

registro = RegistroVehiculos()

while True:
  print("¿Qué acción desea realizar?\n")
  print("- (1) Agregar vehículo ")
  print("- (2) Buscar vehículo ")
  print("- (3) Buscar multas ")
  print("- (4) Buscar dueño ")
  print("- (5)Imprimir certificado ")
  print("- (0)Salir \n")
  opcion = int(input("> "))

  if opcion == 1:
    patente = input("Ingrese la patente del vehículo: ")
    dueno = input("Ingrese el nombre del dueño: ")
    multas = input("Ingrese las multas del vehículo: ")
    certificado = input("Ingrese los detalles del certificado: ")
    vehiculo = Vehiculo(patente, dueno, multas, certificado)
    registro.agregar_vehiculo(vehiculo)
    print("Vehículo agregado con éxito.\n")

  elif opcion == 2:
    patente = input("Ingrese la patente del vehículo: ")
    vehiculo = registro.buscar_vehiculo(patente)
    if vehiculo is not None:
      print(f"La patente del vehículo es {vehiculo.patente}, el dueño es {vehiculo.dueno}, tiene {len(vehiculo.multas)} multas y los detalles del certificado son {vehiculo.certificado}.\n")
    else:
      print("No se encontró el vehículo indicado.\n")

  elif opcion == 3:
    patente = input("Ingrese la patente del vehículo: ")
    multas = registro.buscar_multas(patente)
    if multas is not None:
      print(f"El vehículo con patente {patente} tiene {len(multas)} multas.\n")
    else:
      print(f"No se encontraron multas para el vehículo con patente {patente}.\n")

  elif opcion == 4:
    patente = input("Ingrese la patente del vehículo: ")
    dueno = registro.buscar_dueno(patente)
    if dueno is not None:
      print(f"El dueño del vehículo con patente {patente} es {dueno}.\n")
    else:
      print(f"No se encontró información del dueño para el vehículo con patente {patente}.\n")

  elif opcion == 5:
    patente = input("Ingrese la patente del vehículo: ")
    registro.imprimir_certificado(patente)

  elif opcion == 0:
    print("Adiós!")
    break

  else:
    print("Opción no válida.\n")
```

```python
class AutoSeguro:

 def __init__(self):
        self.registro = {}

def menu(self):
        while True:
            print('Ingrese el número de la opción que desea seleccionar:')
            print('1. Grabar')
            print('2. Buscar')
            print('3. Imprimir certificados')
            print('4. Salir')
            opcion = input()
            if opcion == '1':
                self.grabar_auto()
            elif opcion == '2':
                self.buscar_auto()
            elif opcion == '3':
                self.imprimir_certificados()
            elif opcion == '4':
                print('¡Hasta luego! Gracias por usar Auto Seguro.')
                break
            else:
                print('Opción inválida. Por favor, ingrese el número correspondiente a la opción que desea.')

def grabar_auto(self):
        tipo = input('Ingrese el tipo de vehículo: ')
        while True:
            patente = input('Ingrese la patente del vehículo (sin guiones): ')
            if patente in self.registro:
                print('La patente ya está registrada. Ingrese una patente diferente.')
            else:
                break
        while True:
            marca = input('Ingrese la marca del vehículo (entre 2 y 15 caracteres): ')
            if len(marca) < 2 or len(marca) > 15:
                print('La marca debe tener entre 2 y 15 caracteres. Ingrese una marca válida.')
            else:
                break
        while True:
            precio = input('Ingrese el precio del vehículo: ')
            try:
                precio = int(precio)
                if precio <= 5000000:
                    print('El precio debe ser mayor a $5.000.000. Ingrese un precio válido.')
                else:
                    break
            except ValueError:
                print('El precio debe ser un número entero. Ingrese un precio válido.')
        multas = []
        while True:
            opcion = input('¿Desea agregar una multa? [S/N]')
            if opcion.lower() == 'n':
                break
            monto = input('Ingrese el monto de la multa: ')
            fecha = input('Ingrese la fecha de la multa (en formato dd/mm/aaaa): ')
            multas.append({'monto': monto, 'fecha': fecha})
        fecha_registro = input('Ingrese la fecha de registro del vehículo (en formato dd/mm/aaaa): ')
        nombre_dueño = input('Ingrese el nombre del dueño del vehículo: ')
        self.registro[patente] = {'tipo': tipo, 'marca': marca, 'precio': precio, 'multas': multas, 'fecha_registro': fecha_registro, 'dueño': nombre_dueño}
        print('El vehículo ha sido registrado correctamente.')

def buscar_auto(self):
        patente = input('Ingrese la patente del vehículo que desea buscar: ')
        if patente in self.registro:
            auto = self.registro[patente]
            print(f'Tipo: {auto["tipo"]}')
            print(f'Patente: {patente[:2]}-{patente[2:4]}-{patente[4:]}')
            print(f'Marca: {auto["marca"]}')
            print(f'Precio: ${auto["precio"]:,}')
            print('Multas:')
            for multa in auto['multas']:
                print(f' - Monto: ${multa["monto"]}, Fecha: {multa["fecha"]}')
            print(f'Fecha de registro: {auto["fecha_registro"]}')
            print(f'Dueño actual: {auto["dueño"]}')
        else:
            print(f'No se encontró ningún vehículo con la patente {patente}.')

def imprimir_certificados(self):
        for patente, auto in self.registro.items():
            valores_aleatorios = {'Emisión de contaminantes': 1500, 'Anotaciones vigentes': 2000, 'Multas': 3500}
            for certificado, valor in valores_aleatorios.items():
                print(f'Certificado de {certificado} para el vehículo {patente[:2]}-{patente[2:4]}-{patente[4:]} ({auto["dueño"]}): ${valor:,}')
```
